## ExtraÃ§Ã£o
> #### **saprk.read**

```python
df_bronze = spark.read.format("delta").load(bronzePath)
```

##NormalizaÃ§Ã£o
> #### **datatype**

```python
def up(s): return F.upper(F.trim(F.col(s).cast(T.StringType())))

df_norm = (
    df_bronze
    .withColumn("CO_ISIC_CLASSE", up("CO_ISIC_CLASSE"))
    .withColumn("NO_ISIC_CLASSE", F.col("NO_ISIC_CLASSE").cast(T.StringType()))
    .withColumn("NO_ISIC_CLASSE_ING", F.col("NO_ISIC_CLASSE_ING").cast(T.StringType()))
    .withColumn("NO_ISIC_CLASSE_ESP", F.col("NO_ISIC_CLASSE_ESP").cast(T.StringType()))
    .withColumn("CO_ISIC_GRUPO", up("CO_ISIC_GRUPO"))
    .withColumn("NO_ISIC_GRUPO", F.col("NO_ISIC_GRUPO").cast(T.StringType()))
    .withColumn("NO_ISIC_GRUPO_ING", F.col("NO_ISIC_GRUPO_ING").cast(T.StringType()))
    .withColumn("NO_ISIC_GRURO_ESP", F.col("NO_ISIC_GRUPO_ESP").cast(T.StringType()))
    .withColumn("CO_ISIC_DIVISAO", up("CO_ISIC_DIVISAO"))
    .withColumn("NO_ISIC_DIVISAO", F.col("NO_ISIC_DIVISAO").cast(T.StringType()))
    .withColumn("NO_ISIC_DIVISAO_ING", F.col("NO_ISIC_DIVISAO_ING").cast(T.StringType()))
    .withColumn("NO_ISIC_DIVISAO_ESP", F.col("NO_ISIC_DIVISAO_ESP").cast(T.StringType()))
    .withColumn("CO_ISIC_SECAO", up("CO_ISIC_SECAO"))
    .withColumn("NO_ISIC_SECAO", F.col("NO_ISIC_SECAO").cast(T.StringType()))
    .withColumn("NO_ISIC_SECAO_ING", F.col("NO_ISIC_SECAO_ING").cast(T.StringType()))
    .withColumn("NO_ISIC_SECAO_ESP", F.col("NO_ISIC_SECAO_ESP").cast(T.StringType()))
    .withColumn("TS_REF", F.current_timestamp())
    .withColumn("NM_ORIGEM", F.lit("/landingbeca2026jan/comercio_ext_indices/NCM_ISIC_delta"))
)
```

##ValidaÃ§Ãµes
> #### **data quality**
> #### **deduplicaÃ§Ã£o**
> #### **schema fit**

```python
df_valid = df_norm.filter(F.col("CO_ISIC_CLASSE").isNotNull())
```

```python
df_dedup = df_valid.dropDuplicates(["CO_ISIC_CLASSE"])
```

```python
df_silver = df_dedup.select(
    "CO_ISIC_CLASSE","NO_ISIC_CLASSE","NO_ISIC_CLASSE_ING","NO_ISIC_CLASSE_ESP",
    "CO_ISIC_GRUPO","NO_ISIC_GRUPO","NO_ISIC_GRUPO_ING","NO_ISIC_GRUPO_ESP",
    "CO_ISIC_DIVISAO","NO_ISIC_DIVISAO","NO_ISIC_DIVISAO_ING","NO_ISIC_DIVISAO_ESP",
    "CO_ISIC_SECAO","NO_ISIC_SECAO","NO_ISIC_SECAO_ING","NO_ISIC_SECAO_ESP",
    "TS_REF","NM_ORIGEM"
)
```

##Carga
> #### **merge**

```python
delta_target = DeltaTable.forName(spark, "silver_comercio_ext_indices.tb_referencia_ncm_isic")

merge_condition = """
  t.CO_ISIC_CLASSE = s.CO_ISIC_CLASSE 
"""

(delta_target.alias("t")
    .merge(df_silver.alias("s"), merge_condition)
    .whenMatchedUpdate(set={
        "QT_ESTAT":   "s.QT_ESTAT",
        "KG_LIQUIDO": "s.KG_LIQUIDO",
        "VL_FOB":     "s.VL_FOB",
        "TS_REF":     "s.TS_REF",
        "NM_ORIGEM":  "s.NM_ORIGEM",
    })
    .whenNotMatchedInsertAll()
    .execute()
)
```
