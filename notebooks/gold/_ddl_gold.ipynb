# Tabelas gold

```sql
--DROP DATABASE gold CASCADE;

CREATE DATABASE IF NOT EXISTS gold
LOCATION '/mnt/gold/';
```

```sql
CREATE TABLE IF NOT EXISTS gold.rst_comercio_exterior (
  an_operacao INT,
  me_operacao INT,
  sg_unidade_federativa STRING,
  cd_classificacao_internacional STRING,
  tp_operacao STRING,
  vl_free_on_board FLOAT,
  qt_peso_liquido_kg DECIMAL(20,3)
)
USING DELTA
PARTITIONED BY (an_operacao, me_operacao)
LOCATION '/mnt/gold/rst_comercio_exterior'
TBLPROPERTIES (
  delta.autoOptimize.optimizeWrite = true,
  delta.autoOptimize.autoCompact = true
);
```

```sql
CREATE TABLE IF NOT EXISTS gold.rst_base_empresarial (
  sg_unidade_federativa STRING,
  cod_cnae STRING,
  porte_empresa STRING,
  empresas_ativas LONG,
  empresas_matriz LONG,
  capital_social_total DECIMAL(20,2)
)
USING DELTA
LOCATION 'abfss://gold@storagedatanexus.dfs.core.windows.net/rst_base_empresarial'
TBLPROPERTIES (
  delta.autoOptimize.optimizeWrite = true,
  delta.autoOptimize.autoCompact = true
);
```

```sql
CREATE TABLE IF NOT EXISTS gold.rst_exp_imp_pais (
  ano_operacao INT,
  mes_operacao INT,
  ano_mes STRING,
  cod_pais STRING,
  cod_ncm STRING,
  cod_isic STRING,
  cod_div_isic STRING,
  total_exportado FLOAT,
  total_importado FLOAT,
  saldo FLOAT
)

USING DELTA
PARTITIONED BY (ano_operacao, mes_operacao)
LOCATION 'abfss://gold@storagedatanexus.dfs.core.windows.net/rst_exp_imp_pais'
TBLPROPERTIES (
  delta.autoOptimize.optimizeWrite = true,
  delta.autoOptimize.autoCompact = true
);

DESCRIBE TABLE gold.rst_exp_imp_pais
```

```sql
CREATE TABLE IF NOT EXISTS gold.dim_ncm_isic (
  cod_ncm STRING,
  cod_isic STRING,
  cod_div_isic STRING,
  nome_produto STRING,
  div_isic STRING,
  setor_produto STRING
)

USING DELTA
LOCATION 'abfss://gold@storagedatanexus.dfs.core.windows.net/dim_ncm_isic'
TBLPROPERTIES (
  delta.autoOptimize.optimizeWrite = true,
  delta.autoOptimize.autoCompact = true
);

ALTER TABLE gold.dim_ncm_isic ADD COLUMNS (
  nome_produto STRING,
  setor_produto 
);

DESCRIBE TABLE gold.dim_ncm_isic;
```
